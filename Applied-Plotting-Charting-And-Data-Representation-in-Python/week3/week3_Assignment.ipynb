In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

*Note: The data given for this assignment is not the same as the data used in the article and as a result the visualizations may look a little different.*

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
%matplotlib notebook

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])

# confidence_interval_1 = stats.norm.interval(0.95, loc=32000, scale=2000)
# confidence_interval_2 = stats.norm.interval(0.95, loc=43000, scale=1000)
# confidence_interval_3 = stats.norm.interval(0.95, loc=43500, scale=1400)
# confidence_interval_4 = stats.norm.interval(0.95, loc=48000, scale=700)

#Get year average (year_avg), year standard devition (year_std), number of year (year_count) and standard error (yerr):
year_avg = df.mean(axis = 1)
year_std = df.std(axis = 1)
year_count = df.count(axis = 1)
yerr = year_std*1.96/np.sqrt(year_count)

#Get the lower and upper bound error value in the plot coordinates:
lower_bound = year_avg - yerr
upper_bound = year_avg + yerr
```

**Easiest option implementation:** Implement the bar coloring as described above - a color scale with only three colors, (e.g. blue, white, and red). Assume the user provides the y axis value of interest as a parameter or variable:

```python
#Initialize figure and axes:
fig, ax = plt.subplots()

#Set thereshold (manual input):
threshold = 43000

#Set the bar colour:
colour_list=[]
for l,u in zip(lower_bound,upper_bound):
    if threshold < l:
        colour_list.append("blue")
    elif threshold > u:
        colour_list.append("red")
    else:
        colour_list.append("white")

def plot(colour_list):
    #Set plotting:   
    plot = ax.bar(df.index, year_avg, align='center',yerr= yerr,color = colour_list, edgecolor ="black")

    #set thereshold line as a 2nd y axes and keep the thereshold value:
    ax2 = ax.twinx() 
    ax2.set_ylim(ax.get_ylim())                           
    lhor = ax2.axhline(y = threshold, zorder = 1, color = 'black') 
    ax2.tick_params(axis = 'both', which = 'both', length = 0) 
    ytick = plt.gca().get_yticks()
    ytick = []
    ytick = np.append(ytick,threshold)
    plt.gca().set_yticks(ytick)

    # Add xticks, titles and remove left and top edges:
    plt.xticks(df.index, ('1992', '1993', '1994', '1995'))
    plt.title('Data between 1992 and 1995')
    ax_list = fig.axes # list of axis
    for i in ax_list:
        [i.spines[loc].set_visible(False) for loc in ['top', 'right']]
    
    #return plot, lhor

plot(colour_list)

#show figure:
plt.show()
```

**Harder option implementation:** Implement the bar coloring as described in the paper, where the color of the bar is actually based on the amount of data covered (e.g. a gradient ranging from dark blue for the distribution being certainly below this y-axis, to white if the value is certainly contained, to dark red if the value is certainly not contained as the distribution is above the axis).

```python
#Initialize figure and axes:
fig, ax = plt.subplots()

#Set thereshold (manual input):
threshold = 41000

#Set the bar colour:
colormap = matplotlib.colors.LinearSegmentedColormap.from_list("colormap",["blue", "white", "red"])
def colour_list(threshold):
    colour_list=[]
    for l,u in zip(lower_bound,upper_bound):
        cnorm = matplotlib.colors.Normalize(vmin=l,vmax=u) # normalize based on l and u
        cpick = matplotlib.cm.ScalarMappable(norm=cnorm,cmap=colormap) # scale it based on norm and cmap
        cpick.set_array([]) # set an array
        colour_list.append(cpick.to_rgba(threshold)) #transform it to colour and append it to norm
    return colour_list

#plot
plot(colour_list(threshold))
    
#show figure:
plt.show()
```

**Even Harder option implementation:** Add interactivity to the above, which allows the user to click on the y axis to set the value of interest. The bar colors should change with respect to what value the user has selected.

```python
#Comment 22/08/2002: could net use the "plot" function because cannot control/update the artist with the event handler.

#Initialize figure and axes:
fig, ax = plt.subplots()

#Initializate threshold
threshold = 41000

# #bar plot           
bar_plot = ax.bar(df.index, year_avg, align='center',yerr=yerr,color = colour_list(threshold), edgecolor ="black")

#set line and thereshold in the graph:
ax2 = ax.twinx() # 
ax2.set_ylim(ax.get_ylim()) # set the y axis of ax = to y axis of ax2
ax2.tick_params(axis= 'both', which= 'both',length=0) # remove the ticket length
lhor = ax2.axhline(y = threshold, zorder=1, color = 'black') # set the horizontal line
ytick = plt.gca().get_yticks()
ytick = []
ytick = np.append(ytick,threshold)
plt.gca().set_yticks(ytick)

# Add xticks, titles and remove left and top edges:
plt.xticks(df.index, ('1992', '1993', '1994', '1995'))
plt.title('Data between 1992 and 1995')
ax_list = fig.axes # list of axis
for i in ax_list:
    [i.spines[loc].set_visible(False) for loc in ['top', 'right']]

#event handler
def onclick(event):
    threshold = event.ydata#get threshold based on the event
    lhor.set_ydata(threshold)#update horizontal line
    ytick = ax2.get_yticks()#set ax2 for new thereshold
    ytick = []#set ax2 for new thereshold
    ytick = np.append(ytick,threshold)#set ax2 for new thereshold
    ax2.set_yticks(ytick) #update ax2
    for ind,bar in enumerate(bar_plot):
        bar.set_facecolor(colour_list(threshold)[ind])#update colours for each bar

cid = fig.canvas.mpl_connect('button_press_event', onclick)

#show figure:
plt.show()
```
